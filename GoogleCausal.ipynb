```python
import yfinance as yf
import datetime
```

```python
start= '2020-12-28'
training_end= '2023-03-06'
treatment_start= '2023-03-13'
end = '2023-07-10'
end_stock= '2023-07-11'
```

```python
#get stock data
stocks= ['Meta', 'WMT','DIS', 'BMW.DE','NVS','GS', 'GE','KHC', 'MCD', 'CARL-B.CO']
df= yf.download(stocks,
                start=start,
                end= end_stock,
                interval= '1wk')
```

Extracting the data we only want

```python
df1= df.iloc[:, :10]
```

Getting rid of unneeded level

```python
df1= df1.dropna()
```

Create dataset with most correlated stocks

```python
final_series = df.loc[:, ('Adj Close', ['BMW.DE', 'CARL-B.CO', 'DIS', 'GE', 'MCD', 'META'])]
```

```python
pre_period = [start, training_end]
post_period= [treatment_start, end]
```

```python
!pip install pycausalimpact
```

```python
from causalimpact import CausalImpact
```

```python
impact= CausalImpact(data= final_series,
                     pre_period=pre_period,
                     post_period= post_period,
                     nseasons= [{'period': 52}],
                     period_level_sd= None)
```

```python
impact.plot()
```

```python
print(impact.summary())
```

```python
print(impact.summary('report'))
```
