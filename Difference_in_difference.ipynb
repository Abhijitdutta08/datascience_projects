***Replacing NA's with averages***



```python
import sklearn
```

```python
from sklearn.impute import SimpleImputer
missingvalues= SimpleImputer(missing_values = np.nan,
                             strategy="mean")
```

```python
missingvalues= missingvalues.fit(df[['fte','demp']])
```

```python
df[['fte','demp']]= missingvalues.transform(df[['fte','demp']])
```

```python
df.isnull().any()
```

**Isolating the X and Y variables**

```python
X= df.iloc[:, 0:3]
X
```

```python
Y= df.iloc[:, 3]
Y
```

***Creating the first model***

```python
import statsmodels.api as sm
```

```python
X= sm.add_constant(X)
model1= sm.OLS(Y,X).fit()
```

```python
model1
```

```python
model1.summary(yname = 'Fte',
               xname= ('intercept', 'NJ',	'POST_APRIL92',	'NJ_POST_APRIL92'))
```
